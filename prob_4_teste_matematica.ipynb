```python
from discreteMarkovChain import markovChain
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_prob_deuce(probabilidade: float) -> float:
    if not probabilidade: raise ValueError("Insira a probabilidade!")

    p = probabilidade
    
    P = np.array([
        [0, 1, 0, 0, 0],
        [0, 0, p, 1-p, 0],
        [0, 0, 0, p, 1-p],
        [0, 0, p, 0, 1-p],
        [0, 0, p, 1-p, 0]
    ])
    
    mc = markovChain(P)
    mc.computePi('linear')
    
    return mc.pi[2]
```

```python
probabilidade = 0.5

prob_deuce = calcular_prob_deuce(probabilidade)

print(f"A probabilidade de ocorrer um deuce é aproximadamente {prob_deuce:.4f}")
```
